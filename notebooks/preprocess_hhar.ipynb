# Preprocessing for HHAR dataset

```python
import os
import shutil
import urllib
import zipfile
import warnings
import seaborn as sns
import matplotlib as plt
import pandas as pd
```

```python
warnings.filterwarnings('ignore')
sns.set_style('whitegrid')
```

```python
URL_UCI_HHAR = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00344/Activity recognition exp.zip'
DATASETS_DIR = '../datasets/'
```

```python
def download_hhar_dataset(directory):
    if not os.path.exists(directory + 'HHAR.zip'):
        url = urllib.parse.quote(URL_UCI_HHAR, safe=':/')
        urllib.request.urlretrieve(url, directory + 'HHAR.zip')
    
    if not os.path.exists(directory + 'HHAR'):
        with zipfile.ZipFile(directory + 'HHAR.zip') as zf:
            zf.extractall(path=directory)

        shutil.rmtree(directory + '__MACOSX')
        shutil.move(directory + 'Activity recognition exp',
                    directory + 'HHAR')
```

```python
download_hhar_dataset(DATASETS_DIR)
```
